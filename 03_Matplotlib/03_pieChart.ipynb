```python
import matplotlib.pyplot as plt
```

```python
labels = ["A", "B", "C", "D"]
colors = ["red", "yellow", "orange", "pink"]
sizes = [2,4,6,8]
explode =[0.1, 0.2, 0, 0]
plt.title("split among classes")
plt.pie(sizes, colors = colors, explode = explode, labels = labels, autopct="%.2f%%")
plt.axis("equal")
plt.show()
```
