## Logistic Regression
*Def* - Logistic regression is a classification algorithm used to assign observations to a discrete set of classes.

```python
!pip install scikit-learn
```

```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
cances_ds = datasets.load_breast_cancer()
cances_ds
```

```python
clf = LogisticRegression(C=2)
clf.fit(cances_ds.data, cances_ds.target)
```

```python
clf.predict(cances_ds.data)
```

```python
clf.score(cances_ds.data, cances_ds.target)
```

```python
# here `score` is accuracy
```

```python
clf.predict(cances_ds.data) - cances_ds.target
```

```python
clf.predict_proba(cances_ds.data)
```

## Function which are useful in Logistic REgression

- fit()
- predict()
- score()
- predict_proba()
