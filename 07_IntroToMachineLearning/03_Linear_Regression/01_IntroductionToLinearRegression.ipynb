# Linear Regression


    1. Function which are used in Regular Regression is known as Hypothesis.    
    - Linear Regression is the most basic algorithm in Machine Learning in which we er required to predict continuous values.
    
    Linear Regression -> is assuming Linear relation between X & Y
    
    Terms Related Linear Regression
        1. Features - Independent variables of datasets(x1, x2, x3, ...)
        2. Traget\Output Variable - variable which are depends on features by a relation.
        3. Function\Hypothesis of linear Regression -
                y = m1.x1 + m2.x2 + m3.x3 + ... + mn.xn + c
                
            Hypothesis is a function that tries to fit the data 
        4. Intercept - c is intercept  of the line.
        5. Training Data - data contains features for machine get trained on some hypothesis.

```python
import numpy as np
```

```python
data = np.loadtxt("./data.csv", delimiter=",")
```

```python
data.shape
```

```python
x  = data[:, 0].reshape(-1,1)
y = data[:, 1]
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
alg = LinearRegression()
alg.fit(X_train, Y_train)
```

```python
alg.coef_
```

```python
alg.intercept_
```

```python
import matplotlib.pyplot as plt
m = alg.coef_[0]
c = alg.intercept_

x_line = np.arange(30,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r")
train_1d = X_train.reshape(75)
plt.scatter(X_train, Y_train)
plt.show()
```

```python
x_line = np.arange(30,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r")
test_1d = X_test.reshape(25)
plt.scatter(X_test, Y_test)
plt.show()
```

```python
score_test = alg.score(X_test, Y_test)
score_train = alg.score(X_train, Y_train)
print(score_test, score_train)
```
