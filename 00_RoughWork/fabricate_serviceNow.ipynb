```python
from bs4 import BeautifulSoup
def fabricate_csv(df : "pandas.core.frame.DataFrame"):
    # print(df["number"][0])
    for i in range(0, len(df), 1):
        df['text'][i] = BeautifulSoup(df['text'][i], "lxml").text
    del df['Unnamed: 0']
    return df
    
    
```

```python
import pandas as pd
df1 = pd.read_csv("ouy.csv")
fabricate_csv(df=df1)
```
